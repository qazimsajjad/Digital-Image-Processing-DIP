```python
import cv2
img = cv2.imread('./Fig0236(a)(letter_T).tif',0)
k = np.array([
    [1,1,1],
    [0,0,0],
    [-1,-1,-1]
])
result = np.zeros((img.shape[0], img.shape[1]))
img2 = np.pad(img, (1,1), 'constant')
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        result[i,j] = (img2[i:i+3, j:j+3]*k).sum()
cv2.imshow('Orignal image', img)
cv2.imshow('Resulttant', result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('./Fig1016(a)(building_original).tif', 0)
ker = np.array([
    [1,1,1],
    [1,-8,1],
    [1,1,1]
])
img2 = np.pad(img,(1,1))
result = np.zeros((img.shape[0], img.shape[1]))

for row in range(img.shape[0]-1):
    for col in range(img.shape[1]-1):
        result[row, col]=ker[0,0]*img2[row, col] + ker[0,1]*img2[row,col+1]+ ker[0,2]*img2[row,col+2]\
        +ker[1,0]*img2[row+1,col] + ker[1,1]*img2[row+1, col+1] + ker[1,2]*img2[row+1, col+2]\
        +ker[2,0]*img2[row+2, col] + ker[2,1]*img2[row+2,col+1] + ker[2,2]*img2[row+2, col+2]
cv2.imshow('Original', img)
cv2.imshow('result', result)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
img = cv2.imread('./Fig1016(a)(building_original).tif', 0)
blur = cv2.GaussianBlur(img, (5,5), 0)
ker = np.array([
    [1,1,1],
    [1,-8,1],
    [1,1,1]
])
img2 = np.pad(blur,(1,1))
result = np.zeros((blur.shape[0], blur.shape[1]))

for row in range(blur.shape[0]-1):
    for col in range(blur.shape[1]-1):
        result[row, col]=ker[0,0]*img2[row, col] + ker[0,1]*img2[row,col+1]+ ker[0,2]*img2[row,col+2]\
        +ker[1,0]*img2[row+1,col] + ker[1,1]*img2[row+1, col+1] + ker[1,2]*img2[row+1, col+2]\
        +ker[2,0]*img2[row+2, col] + ker[2,1]*img2[row+2,col+1] + ker[2,2]*img2[row+2, col+2]
cv2.imshow('Image', img)
cv2.imshow('Blurred', blur)
cv2.imshow('Result', result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
