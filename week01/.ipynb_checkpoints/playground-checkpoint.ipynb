```python
import numpy as np
import matplotlib as mt
import matplotlib.pyplot as plt
import cv2
```

```python
#read image
img = cv2.imread('./Fig0241(c)(einstein high contrast).tif',0)
print(img.shape)
result = np.zeros((img.shape[0]//2,img.shape[1]//2), dtype='float32')

for i in range(img.shape[0]-1): #i=0
    for j in range(img.shape[1]-1): #j=1
        result[i//2,j//2] = img[i,j]
        
result = result.astype('int32')
plt.imshow(img, cmap='gray'); 
plt.title('1024');
plt.show()

plt.imshow(result, cmap='gray'); 
plt.title(result.shape);
plt.show()
```
