### Image resizing

```python
img = cv2.imread('../../rose.tif', 0)
result512 = np.zeros((img.shape[0]//2, img.shape[1]//2), dtype=np.uint8)
result256 = np.zeros((img.shape[0]//4, img.shape[1]//4), dtype= np.uint8)
result128 = np.zeros((img.shape[0]//8, img.shape[1]//8), dtype= np.uint8)
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if i % 2 == 0 and j % 2==0:
            result512[i//2,j//2] = img[i,j]
            result256[i//4, j//4] = img[i,j]
            result128[i//8, j//8] = img[i,j]
            
# create figure
fig = plt.figure(figsize=(10, 7))
  
# setting values to rows and column variables
rows = 2
columns = 2

# Adds a subplot at the 1st position
fig.add_subplot(rows, columns, 1)
shape = img.shape
# showing image
plt.imshow(img,cmap = 'gray')
plt.axis('off')
plt.title('Original image')

# Adds a subplot at the 2nd position
fig.add_subplot(rows, columns, 2)

shape = result512.shape
# showing image
plt.imshow(result512, cmap = 'gray')
plt.axis('off')
plt.title('result512')

# Adds a subplot at the 3rd position
fig.add_subplot(rows, columns, 3)
  
# showing image
plt.imshow(result256,cmap = 'gray')
plt.axis('off')
plt.title('result256')

# Adds a subplot at the 4th position
fig.add_subplot(rows, columns, 4)
  
# showing image
plt.imshow(result128,cmap = 'gray')
plt.axis('off')
plt.title('result128')
plt.show()
```

### Image Quantization

```python
## bit changing
img = cv2.imread('./Fig0219(rose1024).tif',0)
shap1=img.shape
result_7_bit = np.zeros((img.shape[0],img.shape[1]), dtype = ('uint8'))
result_6_bit = np.zeros((img.shape[0],img.shape[1]), dtype = ('uint8'))
result_5_bit = np.zeros((img.shape[0],img.shape[1]), dtype = ('uint8'))
result_4_bit = np.zeros((img.shape[0],img.shape[1]), dtype = ('uint8'))
result_3_bit = np.zeros((img.shape[0],img.shape[1]), dtype = ('uint8'))
result_2_bit = np.zeros((img.shape[0],img.shape[1]), dtype = ('uint8'))
result_1_bit = np.zeros((img.shape[0],img.shape[1]), dtype = ('uint8'))



for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        result_7_bit[i][j] = img[i][j]//2
        result_6_bit[i,j] = img[i,j]//4
        result_5_bit[i,j] = img[i,j]//8
        result_4_bit[i,j] = img[i,j]//16
        result_3_bit[i,j] = img[i,j]//32
        result_2_bit[i,j] = img[i,j]//64
        result_1_bit[i,j] = img[i,j]//128
        

cv2.imshow('7-bit', result_7_bit)
cv2.imshow('6-bit', result_6_bit)
cv2.imshow('5-bit', result_5_bit)
cv2.imshow('4-bit', result_4_bit)
cv2.imshow('3-bit', result_3_bit)
cv2.imshow('2-bit', result_2_bit)
cv2.imshow('1-bit', result_1_bit)
cv2.imshow('Original', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
