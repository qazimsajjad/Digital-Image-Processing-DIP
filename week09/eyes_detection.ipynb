```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def imshow(image, title=None,size=5):
    imgType = image.shape
    plt.figure(figsize=(size,size))
    if len(imgType)>2:
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB), 'gray')
    else:
        plt.imshow(image, 'gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
image = cv2.imread('./Trump.jpg')
imshow(image, 'InputImage')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cascadePath = './haarcascade_eye.xml'
cascade = cv2.CascadeClassifier(cascadePath)
face = cascade.detectMultiScale(gray, scaleFactor=1.05, minNeighbors=5,
                                minSize=(30, 30))
for x,y, w,h in face:
    cv2.rectangle(image, (x,y),(x+w, y+h), (0,0,255), 2)
imshow(image, 'Eyes')
```

```python
image = cv2.imread('./DownlaodImage.jpg')
imshow(image, 'InputImage')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
faceCascadePath = './haarcascade_eye.xml'
faceCascade = cv2.CascadeClassifier(cascadePath)

faces = faceCascade.detectMultiScale(gray, 1.3,5)
for (x,y,w,h) in faces:
    cv2.rectangle(image, (x,y), (x+w, y+h), (127,0,255),2)
imshow(image, 'Eyes')
```
