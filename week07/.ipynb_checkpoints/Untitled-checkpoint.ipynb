```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def show(image, title='None'):
    plt.title(title); plt.axis('off')
    plt.imshow(img,'gray')
    plt.show()
```

```python
img = cv2.imread('./Fig0207(a)(gray level band).tif',0)
assert img is not None, 'image path does not exist.....'
show(img, 'Original')
result = np.zeros((img.shape[0], img.shape[1]))


img2= np.pad(img, (2,2),'constant')

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        result[i,j]= (img2[i:i+5, j:j+5]).sum()/9
    
show(result,'Resultant image')
```

```python
img = cv2.imread('./Fig0226(galaxy_pair_original).tif',0)
show(img, 'Original')
result = np.zeros((img.shape[0], img.shape[1]))
kernel = np.ones((7,7))

img2= np.pad(img, (7,7),'constant')

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        result[i,j]= (img2[i:i+7, j:j+7]*kernel).sum()/49
    
show(result,'Resultant image')
```

```python
img = cv2.imread('./Fig0237(a)(characters test pattern)_POST.tif',0)
show(img, 'Original')
result_we = np.zeros((img.shape[0], img.shape[1]))
kernel = np.array([[1,2,1],
                  [2,4,2],
                  [1,2,1]])

img2= np.pad(img, (1,1),'constant')

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        result_we[i,j]= (img2[i:i+3, j:j+3]*kernel).sum()/9
    
show(result_we,'Resultant image')
```

```python
img = cv2.imread('./Fig0237(a)(characters test pattern)_POST.tif',0)
show(result_we, 'result_weighted')
result = np.zeros((img.shape[0], img.shape[1]))
kernel = np.ones((3,3))

img2= np.pad(img, (1,1),'constant')

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        result[i,j]= (img2[i:i+3, j:j+3]*kernel).sum()/9
    
show(result,'un_wighted result')
```

```python
#median filter

img = cv2.imread('./board.tif',0)
show(img, 'Original')
result = np.zeros((img.shape[0], img.shape[1]))
kernel = np.ones((9,9))
m= kernel.shape[0]//2
img2 = np.pad(img, (m,m), 'constant')

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        temp = img2[i:i+9,j:j+9]
        temp = temp.flatten()
        temp_s = np.sort(temp)
        result[i,j] = temp_s[40]
        
show(result, 'resultant image')
```
